```python
#Logic for sorting the pairs
a = ['a','b','c','d']
pairs = []
for p in range(len(a)):
    s = len(a) -1
    for q in range(s):
        q_a = q+1
        if not(p == q_a) and not(p==len(a)-1) and (p < q_a): 
            grouping = a[p] +'<'+ a[q_a]
            pairs.append(grouping)
print(pairs)
```
